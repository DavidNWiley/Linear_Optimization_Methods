# Principal Component Analysis with the SVD

Principal component analysis (PCA) seeks to reduce high dimensional data entries to lower dimensional components.

```julia
using LinearAlgebra
using CSV
using Plots
```

```julia
# Read in the data. CSV reads the data into a "DataFrame" type.
data = CSV.read("DEFRA_Data.csv")
```

```julia
# Make Data Matrix: A_0
A_0 = zeros(17, 4)
A_0[:,1] = copy(data[:,2])
A_0[:,2] = copy(data[:,3])
A_0[:,3] = copy(data[:,4])
A_0[:,4] = copy(data[:,5])
A_0
```

```julia
# Center the data in A_0. This forms the matrix A.
A = copy(A_0)
for i = 1:17
    # Calculate the mean of the i-th row
    μ = sum(A_0[i,:])/4
    
    # Subtract the mean from each entry in the i-th row
    A[i,1] -= μ
    A[i,2] -= μ
    A[i,3] -= μ
    A[i,4] -= μ
end

# Form the sample covariance matrix S.
S = A*A'/(3)
```

```julia
U, Σ, V = svd(A)
```

```julia
scatter(Σ, legend = false, xaxis = "index i", yaxis = "singular value")
```

```julia
foods = data[:,1]
```

```julia
x1 = (U[:,1]'*A[:,1])/(U[:,1]'*U[:,1])
x2 = (U[:,1]'*A[:,2])/(U[:,1]'*U[:,1])
x3 = (U[:,1]'*A[:,3])/(U[:,1]'*U[:,1])
x4 = (U[:,1]'*A[:,4])/(U[:,1]'*U[:,1])

y1 = (U[:,2]'*A[:,1])/(U[:,2]'*U[:,2])
y2 = (U[:,2]'*A[:,2])/(U[:,2]'*U[:,2])
y3 = (U[:,2]'*A[:,3])/(U[:,2]'*U[:,2])
y4 = (U[:,2]'*A[:,4])/(U[:,2]'*U[:,2]);
```

```julia
countries = ["England", "Wales", "Scotland", "Northern Ireland"]
scatter([(x1,0),(x2,0),(x3,0),(x4,0)], series_annotations = text.([x for x in countries], :bottom), legend=false)
```

```julia
scatter([(x1,y1),(x2,y2),(x3,y3),(x4,y4)], series_annotations = text.([x for x in countries], :bottom), legend=false)
```
